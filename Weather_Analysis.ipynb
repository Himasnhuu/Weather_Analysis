### Instances when Snow was recorded

### All the Instances when "Wind speed is above 24" and "Visibility is 25".

```python
data.head(1)
```

```python
data[(data['Wind Speed_km/h'] > 24) & (data['Visibility_km'] == 25)]
```

### Mean value of each column against each 'Weather Condition'.

```python
data.groupby('Weather_Condition').mean(numeric_only=True).head(5)
```

### minimum and maximum value of each column against each weather condition

```python
data.groupby('Weather_Condition').min()
```

```python
data.groupby('Weather_Condition').max()
```

###  All the records where weather condition is Fog

```python
data[data.Weather_Condition.str.contains('Fog')]
```

```python
data[data.Weather_Condition =='Fog']
```

### All the instances when weather is clear or visibility is above 40

```python
data[(data.Weather_Condition == 'Clear') | (data.Visibility_km > 40)]
```

### All the instances (when weather is clear and relative humidity is greater than 50) or visibility is above 40.

```python
data.head(1)
```

```python
data[((data['Weather_Condition'] == 'Clear') & (data['Rel Hum_%'] > 50)) | (data['Visibility_km'] >40)]
```
